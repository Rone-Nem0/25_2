0.0 Delete the env
   ```fish
   sudo rm -rf .venv
   ```
0.1 Create the virtual environment:
   ```fish
   python3 -m venv .venv
   ```

0.2. Activate the environment (fish shell uses a different script):
   ```fish
   source .venv/bin/activate.fish
   ```

0.3. Confirm activation:
   ```fish
   echo $VIRTUAL_ENV
   ```
   If you see the path to .venv, your environment is active.

0.4. To install the requirements from requirements.txt, run:

   ```fish
   pip install -r requirements.txt
   ```

This will create a `_build/site` directory with your website files.
 
## 2. Create a GitHub repository
- Go to [GitHub](https://github.com/) and create a new repository (public or private).
- Do not add a README, .gitignore, or license (to avoid merge conflicts).
 
## 3. Initialize git and push your site
In your project folder, run:
```fish
git init
git commit -m "first commit"
git branch -M main
git remote add origin https://github.com/Rone-Nem0/25_2.git
git push -u origin main
```

In new upload, run:
```fish
git add . 
git commit -m "Upload: New information"
git push
```


 
## 4. Upload the site to the `gh-pages` branch
You can use the `ghp-import` tool to upload the built site:
```fish
pip install ghp-import
ghp-import -n -p -f _build/site
```
- This will push the contents of `_build/site` to the `gh-pages` branch and publish it.
 
## 5. Enable GitHub Pages
- Go to your repository settings on GitHub.
- Scroll to "Pages".
- Set the source to the `gh-pages` branch and `/ (root)` folder.
- Save.
 
## 6. Access your public site
- After a few minutes, your site will be available at:
  `https://<your-username>.github.io/<your-repo>/`
 
## Notes
- Every time you update your book, rebuild with `myst build` and re-run `ghp-import`.
- You can automate this with GitHub Actions for continuous deployment.
 
For more details, see the [MyST documentation](https://mystmd.org/guide/deploying).
