## PMBOK Phases Flowchart

The PMBOK (Project Management Body of Knowledge) defines five key phases in project management: Initiating, Planning, Executing, Monitoring & Controlling, and Closing. Each phase plays a critical role in ensuring project success.

```mermaid
flowchart TD
    A[Initiating<br>Define project goals, stakeholders, and scope] --> B[Planning<br>Develop roadmap, schedule, resources, and risk management]
    B --> C[Executing<br>Carry out project plan, manage teams, and deliverables]
    C --> D[Monitoring & Controlling<br>Track progress, manage changes, ensure quality]
    D --> E[Closing<br>Finalize deliverables, release resources, document lessons learned]
```

- **Initiating:** Establishes the foundation and purpose of the project.
- **Planning:** Outlines how objectives will be achieved, including timelines and resources.
- **Executing:** Focuses on completing the work defined in the plan.
- **Monitoring & Controlling:** Ensures the project stays on track and adapts to changes.
- **Closing:** Wraps up the project, ensuring all goals are met and documenting outcomes.

```mermaid
flowchart TD
    A[Plannification<br> Define the problem, goal, the acction in the time an the resources will use.] --> B[Doing<br>Make the actions in the plan.]
    B --> C[Verification<br> Make the testing of the actions, if the goals was realizated.]
    C --> D[Action<br> if was achieved then documentation and the closer acttion, if not began again the process.]
    D --> A
```
